```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# Import sklearn.preprocessing.StandardScaler
from sklearn.preprocessing import StandardScaler
# Import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import MinMaxScaler

import utils.helper as uh
```

```python
data_dir = './data'

data_file = os.path.join(data_dir, 'processed_data.csv')

df = pd.read_csv(data_file, index_col=0)
```

```python
df
```

```python
df.columns
```

```python
df_copy = df.copy()
```

```python
scaler = StandardScaler() # default=(0, 1)
numerical = ['min_revenue', 'max_revenue', 'avg_revenue']

df_copy[numerical] = scaler.fit_transform(df_copy[numerical])
```

```python
df_copy
```

```python
df_x = df_copy[['industry', 'city', 'state', 'min_revenue', 'max_revenue', 'avg_revenue']]
df_y = df_copy['avg_salary_estimate']
```

```python
df_x = pd.get_dummies(df_x)
```

```python
# Split the 'features' and 'income' data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df_x, 
                                                    df_y, 
                                                    test_size = 0.2, 
                                                    random_state = 0)
```

```python
# Show the results of the split
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
model = SGDRegressor(max_iter=5000, tol=1e-3)
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
r2_score(y_test, predictions)
```

```python
df_copy = df.groupby(['industry', 'city', 'state'], as_index=False).mean()

scaler = StandardScaler() # default=(0, 1)
numerical = ['min_revenue', 'max_revenue', 'avg_revenue']

df_copy[numerical] = scaler.fit_transform(df_copy[numerical])

df_x = df_copy[['industry', 'city', 'state', 'min_revenue', 'max_revenue', 'avg_revenue']]
df_y = df_copy['avg_salary_estimate']

df_x = pd.get_dummies(df_x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_x, 
                                                    df_y, 
                                                    test_size = 0.2, 
                                                    random_state = 0)

model = SGDRegressor(max_iter=5000, tol=1e-3)

model.fit(X_train, y_train)

predictions = model.predict(X_test)

r2_score(y_test, predictions)
```
