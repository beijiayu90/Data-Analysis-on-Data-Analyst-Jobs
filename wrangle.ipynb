# Data Wrangling

```python
df_cleaned = df.copy()
```

```python
column_names_dict = {
    'Job Title': "job_title",
    'Salary Estimate': "salary_estimate",
    'Job Description': "job_description",
    'Rating': "rating",
    'Company Name': "company_name",
    'Location': "location",
    'Headquarters': "headquaters",
    'Size': "size",
    'Founded': "founded",
    'Type of ownership': "type_of_ownership",
    'Industry': "industry",
    'Sector': "sector",
    'Revenue': "revenue",
    'Competitors': "competitors",
    'Easy Apply': "easy_apply"
}

df_cleaned.rename(columns=column_names_dict, inplace=True)
```

```python
df_cleaned.dropna(inplace=True)
```

```python
df_cleaned.info()
```

```python
df_cleaned['salary_estimate'] = df_cleaned['salary_estimate'].str.split('(').str[0].str.strip()
```

```python
indice_to_remove = df_cleaned.query('salary_estimate == "-1"').index

df_cleaned.drop(indice_to_remove, inplace=True)
```

```python
salary_counts = df_cleaned['salary_estimate'].value_counts()
```

```python
indice_to_remove = df_cleaned.query('rating == "-1"').index

df_cleaned.drop(indice_to_remove, inplace=True)
```

# Data Visualization

```python
current_palette = sns.color_palette("colorblind")

sns.set_palette(current_palette)

base_color=current_palette[0]
```

```python
plt.rcParams['figure.figsize'] = (21, 21)

sns.countplot(y="salary_estimate", data=df_cleaned, color=base_color, order=salary_counts.index)
```

```python
plt.rcParams['figure.figsize'] = (8, 8)

bins = np.arange(1, 5+0.5, 0.5)

plt.hist(df_cleaned['rating'], color=base_color, bins=bins)

plt.xticks(bins)

plt.show()
```

```python
heatmap1_data = pd.pivot_table(df_cleaned, values='rating', 
                     index=['salary_estimate'])
                    

sns.heatmap(heatmap1_data, vmin=4, cmap="vlag")

plt.show()
```

# Predict salary based on industry, location, company revenue

```python
df_new = df_cleaned.copy()
```

## I. Data Wrangling

```python
row_to_remove = df_new.query('industry == "-1"').index

df_new.drop(row_to_remove, inplace=True)
```

```python
row_to_remove = df_new.query('revenue == "Unknown / Non-Applicable"').index

df_new.drop(row_to_remove, inplace=True)
```

```python
df_new['revenue'].unique()
```

```python
df_new['revenue'] = df_new['revenue'].str.split('(').str[0].str.strip().str.replace('$', '')
```

```python
df_new['revenue'].unique()
```

```python
processed_revenue = df_new['revenue'].apply(uh.process_revenue)
```

```python
df_new['min_revenue'], df_new['max_revenue'] = processed_revenue.str[0], processed_revenue.str[1]
```

```python
df_new['salary_estimate'].unique()
```

```python
salary_estimate = df_new['salary_estimate'].str.replace(r'[$|K]', '').str.split('-')
```

```python
df_new['min_salary_estimate'], df_new['max_salary_estimate'] = \
    salary_estimate.str[0].astype(int) * 1000, salary_estimate.str[1].astype(int) * 1000
```

```python
splitted_locations = df_new['location'].str.split(',')

df_new['city'], df_new['state'] = splitted_locations.str[0], splitted_locations.str[1]
```

```python
df_new['avg_salary_estimate'] = sum([df_new['min_salary_estimate'], df_new['max_salary_estimate']]) / 2

df_new['avg_salary_estimate'] = pd.to_numeric(df_new['avg_salary_estimate'], downcast='integer')
```

```python
df_new['avg_revenue'] = sum([df_new['min_revenue'], df_new['max_revenue']]) / 2
```

```python
df_new
```

```python
df_new.columns
```

```python
df_final = df_new[[
    'industry', 
    'city',
    'state',
    'min_revenue', 
    'max_revenue', 
    'min_salary_estimate', 
    'max_salary_estimate',
    'avg_salary_estimate',
    'avg_revenue'
]]
```

```python
df_final.reset_index().drop('index', axis=1, inplace=True)
```

```python
df_final
```

```python
output_csv_filepath = os.path.join(data_dir, "processed_data.csv")

df_final.to_csv (output_csv_filepath, index = True, header = True)
```
